**Evaluate Students Letter_Grade**

```python
new_df = df.copy()

# Creating new column to store average total score for each students
new_df['Avg_T_Score'] = (new_df['math_score'] + new_df['read_score'] + new_df['write_score'])/3

# Grading Scale I got from the internet
def letterGrade(score):
    if 80 <= score: 
        return 'A'
    elif 65 <= score: 
        return 'B'
    elif 55 <= score: 
        return 'C'
    elif 50 <= score: 
        return 'D'
    else: 
        return 'F'

# Apply the letterGrade function to get letter grade for each students
new_df['Letter_Grade'] = new_df['Avg_T_Score'].apply(lambda score: letterGrade(score))
new_df.head()
```
